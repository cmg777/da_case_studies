```python
# finally the merged file
#football_managers_merged = pd.read_csv(data_in + "football_managers_workfile.csv").sort_values(["season", "team"])
football_managers_merged = pd.read_csv("https://osf.io/t6dgh/download").sort_values(["season", "team"])
```

```python
games = football_managers_merged.groupby(["team", "manager_id", "manager_name"]).agg(
    manager_games=("div", "size")
)
points = football_managers_merged.groupby(["team", "manager_id", "manager_name"]).agg(
    manager_points=("points", "sum")
)
```

```python
# default join is by indices which is multiindex group key this time
win_ratio = games.join(points)
win_ratio["manager_win_ratio"] = (
    win_ratio["manager_points"] / win_ratio["manager_games"]
)
win_ratio = win_ratio.sort_values("manager_win_ratio")
```

```python
win_ratio = win_ratio.sort_values(["manager_win_ratio"], ascending=False)
```

```python
top_managers = win_ratio.sort_values(by=["manager_win_ratio"], ascending=False).head(11)
top_managers
```

### Figure 2.1 The most successful managers

```python
# denote caretakers
top_managers["manager_win_ratio0"] = [
    row["manager_win_ratio"] if row["manager_games"] < 18 else pd.np.nan
    for index, row in top_managers.iterrows()
]
top_managers["manager_win_ratio1"] = [
    row["manager_win_ratio"] if row["manager_games"] > 18 else pd.np.nan
    for index, row in top_managers.iterrows()
]
```

```python
# denote caretakers
top_managers["fill"] = [
    "1" if row["manager_games"] < 18 else 0 for index, row in top_managers.iterrows()
]
```

```python
top_managers=top_managers.sort_values(["manager_win_ratio"]).reset_index()
top_managers
```

```python
(
    ggplot(
        top_managers,
        aes(
            x="reorder(manager_name,manager_win_ratio)",
            y="manager_win_ratio",
            fill=[color[3] if x == "1" else color[0] for x in top_managers.fill],
        ),
    )
    + geom_col(show_legend=False)
    + ylab("Average points per game")
    + xlab("Manager name")
    + scale_fill_manual(values=(color[0], color[3]))
    + scale_alpha_manual(values=(0.8, 0.3))
    + scale_y_continuous(expand=(0.01, 0.01), limits=(0, 3), breaks=seq(0, 3.1, 0.3))
    + coord_flip()
    + theme_bw()
)
```
